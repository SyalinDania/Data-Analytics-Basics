#### 2 - Cleaning the data

```python
data.isnull().sum()
```

```python
new_data = data.dropna()
```

```python
new_data.isnull().sum()
```

```python
happiness = new_data.drop(['Positive affect','Negative affect'],axis=1)
```

```python
happiness.head()
```

#### 3 - Relationships Analysis

```python
# Correlation Matrix to analyse the data
```

```python
corelation = happiness.corr()
```

```python
sns.heatmap(corelation,xticklabels=corelation.columns,yticklabels=corelation.columns
           ,annot=True)
```

```python
sns.pairplot(happiness)
```

```python
sns.relplot(x='Social support',y='Life Ladder',hue='year',data=happiness)
```

```python
sns.distplot(happiness['Life Ladder'],bins=5)
```

```python
sns.catplot(x='Perceptions of corruption',kind='box',data=happiness)
```
